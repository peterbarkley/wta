```python
# Plot the median, mean, and +/- 1 sigma of the distribution of solution times for each problem size
# using pandas and matplotlib
# load csv file

import pandas as pd
import matplotlib.pyplot as plt

# read csv file
data = pd.read_csv('n10mvary.csv')

# create problem size column = n*m
data['n'] = data['n'].astype(int)
data['m'] = data['m'].astype(int)
data['n*m'] = data['n'] * data['m']

# drop n and m columns
data = data.drop(['n', 'm'], axis=1)

# print(data.groupby('n*m').mean())
print(data.groupby('n*m').median())
# print(data.groupby('n*m').std())
# print(data.groupby('n*m').quantile(0.25))
# print(data.groupby('n*m').quantile(0.75))
# print(data.groupby('n*m').count())
```

```python
# plot mean, median, and quartiles of solution time t vs. problem size m*10
# clear figure
plt.clf()
plt.figure()
# max
plt.plot(data.groupby('n*m').max(), label='max')
plt.plot(data.groupby('n*m').quantile(0.75), label='75%')
#plt.plot(data.groupby('n*m').mean(), label='mean')
plt.plot(data.groupby('n*m').median(), label='median')
plt.plot(data.groupby('n*m').quantile(0.25), label='25%')
plt.legend()
plt.xlabel('problem size (vars)')
plt.ylabel('solution time (s)')
plt.title('Solution time vs. problem size for n=10')
plt.savefig('n10mvary.png')
plt.show()
```
